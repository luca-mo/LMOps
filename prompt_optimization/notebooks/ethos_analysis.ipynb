```python
import pandas as pd
import numpy as np
from statistics import mean, median, mode, stdev
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
data = pd.read_csv('../data/ethos/Ethos_Dataset_Binary.csv', names=['comment', 'label'], sep=';')

# Tokenize the comments and calculate token length
data['token_length'] = data['comment'].apply(lambda x: len(x.split()))

# Calculate statistical measures
mean_length = mean(data['token_length'])
median_length = median(data['token_length'])
mode_length = mode(data['token_length'])
std_dev_length = stdev(data['token_length'])
quantiles = np.percentile(data['token_length'], [25, 50, 75])

# Display the results
print(f"Mean token length: {mean_length}")
print(f"Median token length: {median_length}")
print(f"Mode token length: {mode_length}")
print(f"Standard Deviation of token length: {std_dev_length}")
print(f"25th percentile: {quantiles[0]}")
print(f"50th percentile (Median): {quantiles[1]}")
print(f"75th percentile: {quantiles[2]}")

# Plotting the boxplot
sns.boxplot(x=data['token_length'])
plt.title('Boxplot of Token Lengths')
plt.xlabel('Token Length')
plt.show()
```

```python
# Calculate the token lengths
data['token_length'] = data['comment'].str.len()

# Set the style of the seaborn plot
sns.set(style="whitegrid")

# Create the boxplot
plt.figure(figsize=(10, 6))  # Set the figure size
boxplot = sns.boxplot(x=data['token_length'])

boxplot.axvline(x=400, color='r', linestyle='--', label='400 Token Cutoff')  # Mark the cutoff point
plt.title('Distribution of Comment Token Lengths')  # Add a title
plt.xlabel('Token Length')  # Label the x-axis
plt.legend()  # Add legend to the plot

# Save plot as png
plt.savefig('token_length_distribution.png', dpi=300)

# Show the plot
plt.show()
```
